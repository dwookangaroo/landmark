```python
import itertools

# 제일가까운 5개 호텔 찾기
def find_shortest_hotel(landmark):
    
    dis = {}
    
    for i, j in hotel_dict.items(): 
             
        dis[i] = haversine(landmark_dict[landmark], j, unit='km')        

    A = {k: v for k, v in sorted(dis.items(), key=lambda item: item[1])}
    hotel = dict(itertools.islice(A.items(), 5))
    
    return hotel
    
for lm in landmark_list:
    print(find_shortest_hotel(lm))
```
